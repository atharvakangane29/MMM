### BELOW 3 CELLS CLEANS THE COLUMNS FOR NULL VALUES (STRING NULLS TO ACTUAL NULL VALUES)

```python
catalog = "marketing-analytics-catalog"
schema = "raw_test"

bad_values = "'NULL','null','Null','NA','na','N/A','n/a','','none','None'"

tables = spark.sql(f"SHOW TABLES IN `{catalog}`.{schema}")

for row in tables.collect():

    table_name = row['tableName']
    full_table = f"`{catalog}`.{schema}.{table_name}"

    print(f"Cleaning table: {full_table}")

    df = spark.table(full_table)

    string_cols = [f.name for f in df.schema.fields if f.dataType.simpleString() == "string"]

    if not string_cols:
        print("No string columns found")
        continue

    set_clause = ",\n".join([
        f"""
        {col} = CASE
            WHEN TRIM({col}) IN ({bad_values})
            THEN NULL
            ELSE {col}
        END
        """ for col in string_cols
    ])

    query = f"""
    UPDATE {full_table}
    SET
    {set_clause}
    """

    spark.sql(query)

print("Schema cleanup completed")
```

### BELOW CELL DOES THE DATATYPE MATCHING

```python
from pyspark.sql.functions import col

tables = [
            ("marketing-analytics-catalog","raw_test","dw_veeva_dbo_call"),
            ("marketing-analytics-catalog","raw_test","dw_datasets_centris_speakerprogrambrandtopic"),
            ("ext-dw-veeva-catalog_prod","dbo_test","users"),
            ("ext-dw-veeva-catalog_prod","marketingcloud_test","individualemailresult"),
            ("marketing-analytics-catalog","raw_test","dw_tableau_dbo_accountallterritories"),
            ("marketing-analytics-catalog","raw_test","dw_veeva_dbo_product"),
            ("marketing-analytics-catalog","raw_test","dw_processeddata_dbo_patientactivityhistorydetailed"),
            ("ext-dw-veeva-catalog_prod","dbo_test","profile"),
            ("marketing-analytics-catalog","raw_test","dw_veeva_dbo_calldetail"),
            ("marketing-analytics-catalog","raw_test","dw_datasets_centris_speakerprogram"),
            ("marketing-analytics-catalog","raw_test","dw_datasets_centris_speakerprogramattendee"),
            ("marketing-analytics-catalog","raw_test","dw_tableau_marketing_brandsegment")
            ]

for catalog, schema, table in tables:

    if table not in datatype_config:
        print(f"Skipping {table} (no datatype config)")
        continue

    table_full = f"`{catalog}`.{schema}.{table}"

    print(f"\nProcessing {table_full}")

    df = spark.table(table_full)

    # case-insensitive mapping of actual columns
    actual_cols = {c.lower(): c for c in df.columns}

    dtype_map = datatype_config[table]

    df_casted = df

    for config_col, dtype in dtype_map.items():

        if config_col.lower() in actual_cols:

            real_col = actual_cols[config_col.lower()]

            df_casted = df_casted.withColumn(
                real_col,
                col(real_col).cast(dtype)
            )

        else:
            print(f"⚠️ Column missing in {table_full}: {config_col}")

    (
        df_casted.write
        .mode("overwrite")
        .option("overwriteSchema", "true")
        .saveAsTable(table_full)
    )

    print(f"✅ Datatypes aligned → {table_full}")
```
